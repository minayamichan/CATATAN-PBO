        - SUPER INHERITANCE
          type ini dapat digunakan jika pada konstruktor atau (__init__) terdapat properti yang sama,kita dapat menggunakan
          super().__init__(self) atau menggunakan (nama_kelas).__init__(self).
          Contoh Program :
            
            class Fish:
                    def __init__(self,first_name,last_name="Fish"):
                        self.first_name = first_name
                        self.last_name  = last_name
                    def swim(self):
                        print("The fish is swimming")
                    def swim_backwards(self):
                        print("The fish can swim backward")
            class Trout2(Fish):
                    def __init__(self,water = "Fresh Water"):
                        self.water = water
                        super().__init__(self) #atau dapat menggunakan Fish().__init__(self,first_name,last_name)
            terry = Trout2()
            terry.first_name = "Nemo"
            print(terry.first_name+" "+ terry.last_name)
            print(terry.water)
            terry.swim()

# ENCAPSULATION
1. PENGERTIAN ENCAPSULATION
![](/Github/10a.jpg)

    Encapsulation seperti halnya pil, maksud pil adalah untuk menjaga suatu proses program agar tidak dapat diakses secara sembarangan atau di intervensi oleh program lain. Konsep enkapsulasi sangat penting dilakukan untuk menjaga kebutuhan program agar dapat diakses sewaktu-waktu, sekaligus menjaga program tersebut. Dalam konsep ini ada dua kata kunci yang digunakan, yakni public, private, dan protected :
    * Public memberikan hak akses kepada semua property dan method agar dapat digunakan diluar class 
    * Private memberikan hak akses hanya kepada anggota class yang digunakan untuk mengubah nilai dari property atau method tersebut
    * Protected yaitu memberikan hak akses kepada class itu sendiri dan class turunannya (inheritance), yang ada diclass A hanya bisa diakses oleh class A sendiri.Sedangkan yang tidak mampu mengakses adalah class-class yang berada diluar package class A. untuk dapat mengaksesnya, class yang berada diluar package class A harus meng extends class A.
    * Untuk mengakses method yang terproteksi sendiri ada dua cara yaitu cara menambah _ . (nama kelas) lalu method atau dapat menggunakan metode get/set. Contoh penerapan seperti yang ada dibawah :
