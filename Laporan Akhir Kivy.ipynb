__A. Pengertian Kivy__

Kivy adalah GUI/NUI framework yang menggunakan bahasa pemrograman Python, dapat dijalankan pada Windows, Linux, OS X, Android dan iOS dengan menggunakan code yang sama dan menggunakan MIT license sehingga 100% free to use dan boleh dikomersialisasikan. Kivy didesain untuk membantu pembuatan dan pengembangan aplikasi cross-platform secara cepat.

__B. Cara Instalasi Kivy__

Buat / buka proyek Anda di PyCharm
    
  Di dalam pycharm ada namanya package tambahan yang harus di install agar menunjang keperluan user,untuk menginstall package yang dibutuhkan di dalam pycharm contohnya kivy,sebagai berikut:

1. File -> Pengaturan -> Proyek -> Project Interpreter
2. Klik simbol tambah di sebelah kanan  proyek dan pilih penerjemah yang sesuai

# Penjelasan Class
__1.Class LimitedInput(TextInput)__

Pada _Class LimitedInput(TextInput)_ terdapat 1 method yaitu:

* __insert_text__ = berfungsi untuk memasukkan dan membatasi teks yang akan di input

__2.Class Timer(StackLayout)__

Pada _Class Timer(StackLayout)_ terdapat 4 method:
* __update__ = berfungsi memperbarui total detik yang ada dalam interface
* __clk_start_btn__ = berfungsi menjalankan tombol start yang ada dalam interface
* __clk_reset_timer_btn__ = berfungsi menjalankan tombol ulang/reset pada interface
* __clk_remove_timer_btn__ = berfungsi menjalankan tombol hapus pada interface

__3.Class MainScroll(ScrollView)__

Pada _Class MainScroll(ScrollView)_ hanya terdapat inisialisasi

   * def __init__(self, **kwargs):
         
        super(MainScroll, self).__init__(**kwargs)
        
        self.size_hint = (1, None)
        
        self.size = (Window.width, Window.height)
        
        self.scroll_timeout = 60
        
        self.add_widget(MainScreen())
        
Inisialisasi tersebut berguna untuk gulir utama pada interface

__4.Class MainScreen(StackLayout)__

Pada _Class MainScreen(StackLayout)_ terdapat 4 method :

* __clk_add_timer_btn__ = berfungsi untuk tambahkan tombol timer di interface
* __create_timer__ = berfungsi untuk membuat pengatur waktu pada interface
* __clk_save_and_quit__ = berfungsi untuk tambahkan tombol simpan dan keluar pada interface
* __def save__ = berfungsi untuk menyimpan total detik dan nama timer

__5.Class TimeYourselfApp(App)__

Pada _Class TimeYourselfApp(App)_ hanya terdapat 1 method yaitu:

* __build__ = berfungsi untuk memberi nama aplikasi pada interface
