# 2. INHERITANCE
        - inheritance adalah pewarisan kelas yang dapat menurunkan properti dan method yang dimiliki oleh kelas lain. 
        kelas yang mewarisi (diturunkan) dapat disebut dengan kelas induk kelas ,super class, atau base class.
        sedangkan kelas yang diwarisi (menerima penurunan dari kelas induk) dapat disebut dengan anak kelas (child class).
![](/github/b.jpg)
        Contoh Program :
        
        #Minggu ke 4 | 9/3/2019
        class user:
            nama=""
            def __init__(self,nama):
                self.nama = nama
            def printname(self):
                print ("Nama = ",self.nama)
        class programmer(user):
            def __init__(self,nama):
                self.nama = nama
            def dophyton(self):
                print("Programming phyton")
        brian = user("Brian")
        brian.printname()
        diana = programmer("Diana")
        diana.printname()

        - DEFAULT INHERITANCE
          jika program memiliki method dan properti yang sama, type default ini dapat menggunakan "pass".

            
            Contoh Program :
            class Fish:
                    def __init__(self,first_name,last_name="Fish"):
                        self.first_name = first_name
                        self.last_name  = last_name
                    def swim(self):
                        print("The fish is swimming")
                    def swim_backwards(self):
                        print("The fish can swim backward")
            class Trout(Fish):
                    pass
            class Clownfish(Fish):
                    def live_with_anemone(self):
                    print("The clownfish is coexisting with sea anemone")
            #default
            casey = Clownfish("Casey")
            print(casey.first_name+" "+ casey.last_name)
            casey.swim()
            casey.live_with_anemone()

        - SUPER INHERITANCE
          type ini dapat digunakan jika pada konstruktor atau (__init__) terdapat properti yang sama,kita dapat menggunakan
          super().__init__(self) atau menggunakan (nama_kelas).__init__(self).
          Contoh Program :
            
            class Fish:
                    def __init__(self,first_name,last_name="Fish"):
                        self.first_name = first_name
                        self.last_name  = last_name
                    def swim(self):
                        print("The fish is swimming")
                    def swim_backwards(self):
                        print("The fish can swim backward")
            class Trout2(Fish):
                    def __init__(self,water = "Fresh Water"):
                        self.water = water
                        super().__init__(self) #atau dapat menggunakan Fish().__init__(self,first_name,last_name)
            terry = Trout2()
            terry.first_name = "Nemo"
            print(terry.first_name+" "+ terry.last_name)
            print(terry.water)
            terry.swim()

        - MULTIPLE INHERITANCE
            type ini digunakan untuk menggabungkan kelas 1 dan kelas 2.
          
          Contoh Program :
          
            class coral:
                    def community(self):
                        print("Coral Lives in a community")

            class coralreef(coral,anemone):
                    pass

# POLYMORPHISME
    - Polymorphisme adalah salah satu sifat yang berbeda kelas ( namun tetap terikat inheritance ) untuk memanggil method yang  mempunyai nama yang sama
      
      Contoh Program :

            def swim (self):
                print ("The shark is swimmming")
            def swim_backwards(self):
                print ("The shark cannot swim backwards, but can sink backwards.")
            def skeleton(self):
                print("The shark's skeleton is made of cartinlage.")
     class clownfish():

                print("The clownfish's skeleton is made of bone")
     sammy = shark()

    - Polymorphisme With Fuction
      Polymorphisme yang memanggil program dengan Fuction / Fungsi (for .... in ... ) :
      Contoh Program :
      
      class shark():

        def swim_backwards(self):
            print("The clownfish can swim backwards.")
        def skeleton(self):
            print("The clownfish's skeleton is made of bone")
        def in_the_pacific(fish):
