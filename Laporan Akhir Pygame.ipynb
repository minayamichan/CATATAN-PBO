__A. PENGERTIAN PYGAME__

Pygame adalah seperangkat modul python yang dirancang untuk membuat suatu permainan. Didalam pygame tersebut terdapat fungsi- fungsi tersendiri seperti halnya untuk bagian multimedia, seperti karakter game, sound, dan lain sebagainya.Pygame sendiri menggunkan library SDL yang memungkinkan para pengguna dapat menggunkan berbagai macam fitur dan fungsi yang disediakan oleh pygame.

Pygame sendiri memiliki beberapa keunggulan yakni mudah di mengerti bagi pemula, berlisensi gratis, dan memiliki beberapa komunitas yang dapat membantu dalam belajar.

__B. INSTALASI PYGAME__

Sebelum membuat game, hal yang harus dilakukan pertama kali adalah menginstall modul pygame tersebut terlebih dahulu. Berikut adalah langkah-langkah dalam menginstall modul pygame:

Buka CMD, kemudian ketikkan “pip”
Setelah muncul menu dari pip itu sendiri kemudian ketik “pip install pygame”
Kemudian akan otomatis mendownload file instalasi modul pygame. Jika sudah muncul tulisan “Successfully installed pygame-1.9.6” , maka pygame sudah terinstall.

__C. PENJELASAN CLASS__
      
1. Class Player
Semua class yang digunakan dalam pembuatan game mewarisi class sprite milik pygame

Pada class player terdapat 4 method yang method inisialisasi, method update, method powerup, method shoot. Keempat method tersebut memiliki atribut dan fungsi masing masing. Yaitu :

   __A. Inisialisasi__
   * Sebuah sprite harus mewarisi sprite dari library pygame.
   * Pygame.image.load berfungsi untuk membaca img_dir (tempat menyimpan gambar game) yang digunakan dalam game.
   * Atribut vx dan vy merupakan posisi awal player.
   * Atribut shield adalah jumlah shield yang dimiliki oleh player, jika shield == 0, maka player.lives -= 1.
   * Atribut shoot_delay berfungsi untuk cepat lambatnya shooting pertama dengan shooting kedua.
   * Atribut self.power adalah untuk jumlah peluru yang dimiliki player saat pertama bermain.

   __B. Update__
   * Method ini digunakan untuk mendeteksi tombol keyboard panah (atas,bawah) yang berfungsi untuk menggerakkan player.
   * Tombol keyboard yang ditekan di deteksi dengan menggunakan key.get_pressed().

   __C. Powerup__
   * self.power += 1 digunakan untuk ketika player hit powerup, power player akan bertambah 1.

   __D. Shoot__
   * Method ini mewarisi sifat dari kelas Bullet.
